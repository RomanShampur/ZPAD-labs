```python
import pandas as pd
import requests
import io
import re
import time
from IPython.display import display, HTML

# 1. Назви областей
province_names = {
    1: "Вінницька", 2: "Волинська", 3: "Дніпропетровська", 4: "Донецька", 5: "Житомирська",
    6: "Закарпатська", 7: "Запорізька", 8: "Івано-Франківська", 9: "Київська", 10: "Кіровоградська",
    11: "Луганська", 12: "Львівська", 13: "Миколаївська", 14: "Одеська", 15: "Полтавська",
    16: "Рівненська", 17: "Сумська", 18: "Тернопільська", 19: "Харківська", 20: "Херсонська",
    21: "Хмельницька", 22: "Черкаська", 23: "Чернівецька", 24: "Чернігівська", 25: "Республіка Крим"
}

# 2. Мапа NOAA ID (відповідно до твого скріншоту, де 22 - Тернопільська)
ukr_to_noaa = {
    1: 22, 2: 24, 3: 23, 4: 25, 5: 3, 6: 4, 7: 8, 8: 19, 9: 20, 10: 21,
    11: 9, 12: 13, 13: 14, 14: 15, 15: 16, 16: 17, 17: 18, 18: 6, 19: 1,
    20: 2, 21: 7, 22: 5, 23: 10, 24: 11, 25: 12
}

all_frames = []
headers = {'User-Agent': 'Mozilla/5.0'}

print("Починаю збір даних... Зачекай трохи.")

for ukr_id in range(1, 26):
    noaa_id = ukr_to_noaa[ukr_id]
    url = f"https://www.star.nesdis.noaa.gov/smcd/emb/vci/VH/get_TS_admin.php?country=UKR&provinceID={noaa_id}&year1=1981&year2=2024&type=Mean"
    
    try:
        response = requests.get(url, headers=headers)
        text = response.text
        
        # 1. Видаляємо HTML теги
        clean_text = re.sub(r'<.*?>', '', text)
        
        # 2. Витягуємо лише рядки з даними (починаються з року 19xx або 20xx)
        # Використовуємо rstrip(','), щоб прибрати ту зайву кому в кінці!
        data_lines = []
        for line in clean_text.split('\n'):
            line = line.strip().rstrip(',') 
            if re.match(r'^\d{4}', line):
                data_lines.append(line)
        
        if not data_lines:
            continue
            
        # 3. Створюємо DataFrame з чистих рядків
        df = pd.read_csv(io.StringIO('\n'.join(data_lines)), 
                         names=['Year', 'Week', 'SMN', 'SMT', 'VCI', 'TCI', 'VHI'])
        
        # 4. Чистимо дані
        df = df[df['VHI'] > 0]
        df['Province_ID'] = ukr_id
        df['Province_Name'] = province_names[ukr_id]
        
        all_frames.append(df)
        print(f"Область {ukr_id}: {province_names[ukr_id]} — успішно.")
        
        time.sleep(0.2) # Легка пауза
        
    except Exception as e:
        print(f"Помилка в області {ukr_id}: {e}")

# Вивід результату
if all_frames:
    final_df = pd.concat(all_frames, ignore_index=True)
    final_df.to_csv('final_vhi_data.csv', index=False)
    
    display(HTML("<h2 style='color: green;'>Ура! Дані завантажено!</h2>"))
    display(final_df.head(10))
    print(f"Всього рядків: {len(final_df)}")
else:
    print("Даних немає. Спробуй ще раз.")
```

```python
import pandas as pd
import numpy as np
import timeit
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# 1. Завантаження та очищення даних
def load_and_clean_power_data(filepath):
    # Використовуємо sep=';', оскільки в цьому файлі розділювач - крапка з комою
    # na_values=['?'] - у цьому датасеті пропуски позначені знаком питання
    df = pd.read_csv(filepath, sep=';', low_memory=False, na_values=['?'])
    
    # Видаляємо пропуски (Data Cleaning)
    df = df.dropna()
    
    # Конвертуємо числові колонки у відповідний тип
    float_cols = ['Global_active_power', 'Global_reactive_power', 'Voltage', 
                  'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']
    df[float_cols] = df[float_cols].astype(float)
    
    # Створюємо колонку з часом для зручності фільтрації
    df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.time
    
    return df

df_power = load_and_clean_power_data('household_power_consumption.txt')

# --- ВИБІРКИ ТА ПРОФІЛЮВАННЯ ---

# Завдання 1: Активна потужність > 5 кВт
def task_1(df):
    return df[df['Global_active_power'] > 5]

t1_time = timeit.timeit(lambda: task_1(df_power), number=1)
print(f"Завдання 1 виконано за: {t1_time:.5f} сек. Знайдено записів: {len(task_1(df_power))}")

# Завдання 2: Сила струму 19-20 А + (кухня > бойлер)
def task_2(df):
    # Sub_metering_1 + 2 (пралка/холодильник) > Sub_metering_3 (бойлер/кондер)
    return df[(df['Global_intensity'] >= 19) & (df['Global_intensity'] <= 20) & 
              ((df['Sub_metering_1'] + df['Sub_metering_2']) > df['Sub_metering_3'])]

t2_time = timeit.timeit(lambda: task_2(df_power), number=1)
print(f"Завдання 2 виконано за: {t2_time:.5f} сек. Знайдено записів: {len(task_2(df_power))}")

# Завдання 3: Випадкові 500 000 записів та середні величини
def task_3(df):
    sample = df.sample(n=500000, replace=False)
    means = sample[['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']].mean()
    return means

t3_time = timeit.timeit(lambda: task_3(df_power), number=1)
print(f"Завдання 3 виконано за: {t3_time:.5f} сек.")
print("Середні значення груп споживання:\n", task_3(df_power))

# Завдання 4: Після 18:00, потужність > 6 кВт, група 2 найбільша + специфічний відбір
def task_4(df):
    import datetime
    # Фільтрація за часом та потужністю
    time_limit = datetime.time(18, 0, 0)
    filtered = df[(df['Time'] > time_limit) & (df['Global_active_power'] > 6)]
    
    # Визначаємо, де група 2 (Sub_metering_2) є найбільшою
    res = filtered[(filtered['Sub_metering_2'] > filtered['Sub_metering_1']) & 
                   (filtered['Sub_metering_2'] > filtered['Sub_metering_3'])]
    
    # Відбір кожного 3-го з першої половини та кожного 4-го з другої
    half = len(res) // 2
    first_half = res.iloc[:half:3]
    second_half = res.iloc[half::4]
    
    return pd.concat([first_half, second_half])

t4_time = timeit.timeit(lambda: task_4(df_power), number=1)
print(f"Завдання 4 виконано за: {t4_time:.5f} сек. Результатів: {len(task_4(df_power))}")

# --- ДОДАТКОВІ ОБРОБКИ ---

# 5. Нормалізація та Стандартизація
# Беремо частину даних для демонстрації
scaler_minmax = MinMaxScaler()
scaler_std = StandardScaler()

df_normalized = df_power.copy()
df_normalized[['Global_active_power', 'Voltage']] = scaler_minmax.fit_transform(df_power[['Global_active_power', 'Voltage']])
print("\nДані після нормалізації (MinMax):")
display(df_normalized[['Global_active_power', 'Voltage']].head())

# 6. Коефіцієнти Пірсона та Спірмена
pearson = df_power['Global_active_power'].corr(df_power['Global_intensity'], method='pearson')
spearman = df_power['Global_active_power'].corr(df_power['Global_intensity'], method='spearman')
print(f"\nКореляція Пірсона: {pearson:.4f}")
print(f"Кореляція Спірмена: {spearman:.4f}")

# 7. One Hot Encoding (на прикладі фіктивного категоріального атрибута)
# Оскільки в датасеті немає категорій, створимо їх для демонстрації (наприклад, рівень споживання)
df_power['Usage_Level'] = pd.cut(df_power['Global_active_power'], bins=[0, 2, 5, 20], labels=['Low', 'Medium', 'High'])
df_ohe = pd.get_dummies(df_power, columns=['Usage_Level'])
print("\nРезультат One Hot Encoding (перші 5 рядків):")
display(df_ohe.filter(like='Usage_Level').head())
```
